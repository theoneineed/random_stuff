```python
import matplotlib.pyplot as plt
import math
import numpy as np
i=0
theta_data=np.arange(0,91,1)
ref_data=[]
tra_data=[]
label1=r'$\frac{\tilde{E_{o_{R}}}}{\tilde{E_{o_{I}}}}$'
label2=r'$\frac{\tilde{E_{o_{T}}}}{\tilde{E_{o_{I}}}}$'
for item in theta_data:
    t=i*(np.pi/180)
    ref_data.append((np.cos(t)-np.sqrt(2.25-(np.sin(t))**2))/(np.cos(t)+np.sqrt(2.25-(np.sin(t))**2)))
    tra_data.append((2*np.cos(t))/(np.cos(t)+np.sqrt(2.25-(np.sin(t))**2)))
    #print ('('+str(theta_data[i])+','+str(ref_data[i])+','+str(tra_data[i])+')')
    i=i+1
fig=plt.figure(1,figsize=(10,10))
plt.plot(theta_data,ref_data,color='red',linestyle='-',label=label1)
plt.plot(theta_data,tra_data,color='blue',linestyle='--',label=label2)
plt.legend()
plt.xlabel(r'$\theta$',size=15)
fig.savefig('drawn_python.png',dpi=300,bbox='tight')
```
